```python
#2203A52126 CAPSTON PROJECT STAT ML WITH LinearRegression done Manually
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


data = pd.read_csv('/content/Cellphone.csv')
data.fillna(data.mean(), inplace=True)
features = data.drop('Price', axis=1)
target = data['Price']


numerical_features = [
    'Product_id', 'Sale', 'weight', 'resoloution', 'ppi',
    'cpu freq', 'internal mem', 'ram', 'RearCam', 'Front_Cam',
    'battery', 'thickness'
]


for feature in numerical_features:
    plt.scatter(features[feature], target, c='red', s=50)
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.title(f'{feature} vs Price')
    plt.show()


scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)


X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.3, random_state=13)

def linear_regression(X, y):
    X_transpose = X.T
    coefficients = np.linalg.inv(X_transpose @ X) @ X_transpose @ y
    return coefficients

X_train = np.column_stack((np.ones(X_train.shape[0]), X_train))
y_train = y_train.to_numpy()

coefficients = linear_regression(X_train, y_train)

X_test = np.column_stack((np.ones(X_test.shape[0]), X_test))
y_pred = X_test @ coefficients


equation = f'y = {intercept:.5f}'
for feature, coeff in zip(features.columns, coefficients):
    equation += f' + {coeff:.5f} * {feature}'
print('Equation is :\n',equation)


plt.scatter(y_test, y_pred, c='green', s=20)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='-', color='black', linewidth=2)
plt.show()


residuals = y_test - y_pred
mse = np.mean(residuals**2)
mae = np.mean(np.abs(residuals))
rmse = np.sqrt(mse)
print(f"Mean Squared Error (MSE): {mse:.5f}")
print(f"Mean Absolute Error (MAE): {mae:.5f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.5f}")


p = float(input("Enter Product_id: "))
s = float(input("Enter Sale: "))
w = float(input("Enter weight: "))
r = float(input("Enter resoloution: "))
ppi = float(input("Enter ppi: "))
cpu_c = float(input("Enter cpu core: "))
cpu_f = float(input("Enter cpu freq: "))
intmem = float(input("Enter internal mem: "))
ram = float(input("Enter ram: "))
rearcam = float(input("Enter rear cam: "))
f_cam = float(input("Enter frontcam: "))
battery = float(input("Enter battery: "))
thickness = float(input("Enter thickness: "))


predicted_price = (
    intercept +
    coefficients[1] * p +
    coefficients[2] * s +
    coefficients[3] * w +
    coefficients[4] * r +
    coefficients[5] * ppi +
    coefficients[6] * cpu_c +
    coefficients[7] * cpu_f +
    coefficients[8] * intmem +
    coefficients[9] * ram +
    coefficients[10] * rearcam +
    coefficients[11] * f_cam +
    coefficients[12] * battery +
    coefficients[13] * thickness
)

print(f'Predicted price for your features is: {predicted_price:.5f}')
```

```python
#2203A52126 CAPSTON PROJECT STAT ML WITH LASSO REGRESSION
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error


data = pd.read_csv('/content/Cellphone.csv')
data.fillna(data.mean(), inplace=True)
features = data.drop('Price', axis=1)
target = data['Price']


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)


scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


lasso = Lasso(alpha=0.01)


lasso.fit(X_train, y_train)


y_pred = lasso.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)


coefficients = lasso.coef_
plt.figure(figsize=(10, 5))
plt.bar(features.columns, coefficients)
plt.title("Lasso Regression Coefficients")
plt.xticks(rotation=90)
plt.show()

coefficients = lasso.coef_
intercept = lasso.intercept_


feature_names = features.columns


equation = f'Price = {intercept:.2f}'
for i, feature in enumerate(feature_names):
    equation += f' + {coefficients[i]:.2f} * {feature}'

print("Lasso Regression Equation:")
print(equation)



user_inputs = []
for feature in feature_names:
    user_input = float(input(f"Enter the value for {feature}: "))
    user_inputs.append(user_input)


user_inputs_scaled = scaler.transform([user_inputs])


predicted_price = lasso.predict(user_inputs_scaled)

print(f"Predicted Price: {predicted_price[0]:.2f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
from sklearn.utils import resample
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Cellphone.csv')
data.fillna(data.mean(), inplace=True)
features = data.drop('Price', axis=1)
target = data['Price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize models
linear_reg = LinearRegression()
lasso_reg = Lasso()
svm_reg = SVR()
knn_reg = KNeighborsRegressor()
ridge_reg = Ridge()

# Number of bootstrap iterations
num_iterations = 100

# Lists to store mean squared errors (MSE) for each model
linear_mses = []
lasso_mses = []
svm_mses = []
knn_mses = []
ridge_mses = []

for i in range(num_iterations):
    # Bootstrap resampling
    X_resampled, y_resampled = resample(X_train, y_train, random_state=i)

    # Fit and evaluate Linear Regression
    linear_reg.fit(X_resampled, y_resampled)
    linear_reg_pred = linear_reg.predict(X_test)
    linear_mse = mean_squared_error(y_test, linear_reg_pred)
    linear_mses.append(linear_mse)

    # Fit and evaluate Lasso Regression
    lasso_reg.fit(X_resampled, y_resampled)
    lasso_reg_pred = lasso_reg.predict(X_test)
    lasso_mse = mean_squared_error(y_test, lasso_reg_pred)
    lasso_mses.append(lasso_mse)

    # Fit and evaluate Support Vector Machine (SVM) Regression
    svm_reg.fit(X_resampled, y_resampled)
    svm_reg_pred = svm_reg.predict(X_test)
    svm_mse = mean_squared_error(y_test, svm_reg_pred)
    svm_mses.append(svm_mse)

    # Fit and evaluate K-Nearest Neighbors (KNN) Regression
    knn_reg.fit(X_resampled, y_resampled)
    knn_reg_pred = knn_reg.predict(X_test)
    knn_mse = mean_squared_error(y_test, knn_reg_pred)
    knn_mses.append(knn_mse)

    # Fit and evaluate Ridge Regression
    ridge_reg.fit(X_resampled, y_resampled)
    ridge_reg_pred = ridge_reg.predict(X_test)
    ridge_mse = mean_squared_error(y_test, ridge_reg_pred)
    ridge_mses.append(ridge_mse)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.title("Bootstrap Accuracy vs. Number of Iterations")
plt.xlabel("Iteration")
plt.ylabel("Mean Squared Error")
plt.plot(range(1, num_iterations + 1), linear_mses, label="Linear Regression")
plt.plot(range(1, num_iterations + 1), lasso_mses, label="Lasso Regression")
plt.plot(range(1, num_iterations + 1), svm_mses, label="SVM Regression")
plt.plot(range(1, num_iterations + 1), knn_mses, label="KNN Regression")
plt.plot(range(1, num_iterations + 1), ridge_mses, label="Ridge Regression")
plt.legend()
plt.show()
```
